```python
from llm_runner import TestCase

GD_VALIDITY_URL = "https://search.clinicalgenome.org/kb/gene-validity/download"
```

```python
import pandas as pd
```

```python
df = pd.read_csv(GD_VALIDITY_URL, skiprows=4)
```

```python
# Drop the redundant row with "+++++++++" symbols
df = df[1:].reset_index(drop=True)
```

```python
df.head()
```

```python
# summarize by status
df['CLASSIFICATION'].value_counts()
```

```python
disputed = df[df['CLASSIFICATION'] == 'Disputed']
disputed
```

```python
disputed_genes = disputed['GENE SYMBOL'].unique()
```

```python
definitive = df[df['CLASSIFICATION'] == 'Definitive']
definitive
```

```python
# for each gene, pick one definitive classification and one disputed classification
balanced_df = pd.DataFrame()
for gene in df['GENE SYMBOL'].unique():
    gene_df = df[df['GENE SYMBOL'] == gene]
    definitive_gene = gene_df[gene_df['CLASSIFICATION'] == 'Definitive']
    if len(definitive_gene) == 0:
        continue
    disputed_gene = gene_df[gene_df['CLASSIFICATION'] == 'Disputed']
    if len(disputed_gene) == 0:
        continue
   # Take first row from each and concatenate
    balanced_df = pd.concat([
        balanced_df,
        definitive_gene.head(1),  # Get first row of definitive
        disputed_gene.head(1)     # Get first row of disputed
    ])

    
balanced_df
```

```python
from llm_runner import TestCase

def row_to_case(row: dict):
    return TestCase(
        input=f"Variants in {row['GENE SYMBOL']} are implicated in {row['DISEASE LABEL']}",
        ideal="YES" if row['CLASSIFICATION'] == 'Definitive' else "NO",
        original_input={k: str(v) for k, v in row.items()},
    )
```

```python
cases = [row_to_case(row) for _, row in balanced_df.iterrows()]
```

```python
header = """
name: gene_disease_refuted
template: binary_with_explanation
templates:
  binary_with_explanation:
    system: Is the provided statement true? Return YES, NO, or OTHER, followed by an explanation.
      The explanation can be of any form and length but your response must start with YES, NO, or OTHER.
    prompt: "{input}"
    metrics:
      - qa_with_explanation
models:
  citeseek-gpt-4o:
    plugins:
      - citeseek
    parameters:
      model: gpt-4o
matrix:
  hyperparameters:
    model: [gpt-4o, deepseek-chat, claude-sonnet, gemini-2.0-flash-exp, deepseek-reasoner, citeseek-gpt-4o]
    temperature: [0.0, 0.05]
"""
    
```

```python
from pathlib import Path

path = Path("clingen-validity-tmp.yaml")
```

```python
if False and path.exists():
    raise FileExistsError("File already exists")
```

```python
import yaml
with open(path, "w") as f:
    f.write(header)
    cases_dicts = [case.model_dump(exclude_unset=True) for case in cases]
    yaml.dump({"cases": cases_dicts}, f, sort_keys=False)
```
